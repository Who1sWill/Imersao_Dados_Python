```python
# @title
df.rename(columns={
    'ano_trabalho': 'ano',
    'nivel_experiencia': 'senioridade',
    'tipo_emprego': 'contrato',
    'cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_usd': 'usd',
    'residencia_funcionario': 'residencia',
    'taxa_remoto': 'remoto',
    'local_empresa': 'empresa',
    'tamanho_empresa': 'tamanho_empresa'
}, inplace=True)

df.head()
```

```python
# @title
df["senioridade"].value_counts()
```

```python
# @title
df["contrato"].value_counts()
```

```python
# @title
df["remoto"].value_counts()
```

```python
# @title
df["tamanho_empresa"].value_counts()
```

```python
# @title
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

df['contrato'].value_counts()
```

```python
# @title
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Snior',
    'MI': 'Pleno',
    'EN': 'Jnior',
    'EX': 'Executivo'
})

df['senioridade'].value_counts()
```

```python
# @title
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Mdia',
    'L': 'Grande',
    'S': 'Pequena'
})

df['tamanho_empresa'].value_counts()
```

```python
# @title
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Hbrido'
})

df['remoto'].value_counts()
```

```python
import numpy as np
df_salarios = pd.DataFrame({
    'nome':["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'salario': [4000, np.nan, 5000, np.nan, 100000 ]
})

#calcula a media salarial e substitui os nulos pela mdia e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

''' Calcula a mediana e substitui os nulos pela mediana

'''

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Tera", "Quarta", "Quinta", "Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["Preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Tera", "Quarta", "Quinta", "Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["Preenchido_bfill"] = df_temperaturas["Temperatura"].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome':["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'cidade': ["So Paulo", np.nan, "Curitiba", np.nan, "Belm"]
})

df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("No informado")
df_cidades
```

```python
import seaborn as sns
```

```python
sns.barplot(data=df_limpo, x='senioridade', y='usd')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df_limpo, x='senioridade', y='usd')
plt.title("Salrio mdio por senioridade")
plt.xlabel("Senioridade")
plt.ylabel("Salrio mdio anual (USD)")
plt.show()
```

```python
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=True).index
```

```python
ordem
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df_limpo, x='senioridade', y='usd', order = ordem)
plt.title("Salrio mdio por senioridade")
plt.xlabel("Senioridade")
plt.ylabel("Salrio mdio anual (USD)")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df_limpo['usd'], bins = 50, kde=True)
plt.title("Distribuio de salrios anuais")
plt.xlabel("Salrio em USD")
plt.ylabel("Frequncia")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_limpo['usd'])
plt.title("Bloxplot Salrio")
plt.xlabel("Salrio em USD")
plt.show()
```

```python
ordem_senioridade = ['Jnior', 'Pleno', 'Snior', 'Executivo']
plt.figure(figsize=(8, 5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade)
plt.title("Bloxplot da distribuio salarial por senioridade")
plt.xlabel("Salrio em USD")
plt.show()
```

```python
ordem_senioridade = ['Jnior', 'Pleno', 'Snior', 'Executivo']
plt.figure(figsize=(8, 5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade, palette='Set2', hue='senioridade')
plt.title("Bloxplot da distribuio salarial por senioridade")
plt.xlabel("Salrio em USD")
plt.show()
```

```python
import plotly.express as px
```

```python
df_senioridade_media = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(
    df_senioridade_media,
    x='senioridade',
    y='usd',
    title='Mdia Salarial por Senioridade',
    labels={'senioridade': 'Nvel de Senioridade', 'usd': 'Mdia Salarial Anual (USD)'},
    color='senioridade', # Cor por senioridade
    color_discrete_sequence=px.colors.qualitative.Set2 # Paleta de cores
)

fig.update_layout(xaxis={'categoryorder':'array', 'categoryarray':['Jnior', 'Pleno', 'Snior', 'Executivo']})
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(
    remoto_contagem,
    names='tipo_trabalho', values='quantidade',
    title='Proporo dos tipos de trabalho',
    hole=0.5
)
fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
# Filtrar o DataFrame para o cargo de 'Data Scientist'
df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']

# Agrupar por pas (empresa) e calcular a mdia salarial em USD
df_salario_pais_data_scientist = df_data_scientist.groupby('empresa')['usd'].mean().sort_values(ascending=False).reset_index()

# Criar o grfico de barras interativo usando Plotly Express
fig = px.bar(
    df_salario_pais_data_scientist,
    x='empresa',
    y='usd',
    title='Mdia Salarial para Data Scientists por Pas',
    labels={'empresa': 'Pas da Empresa', 'usd': 'Mdia Salarial Anual (USD)'},
    color='empresa', # Cor por pas
    color_discrete_sequence=px.colors.qualitative.Plotly # Paleta de cores
)

fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```
